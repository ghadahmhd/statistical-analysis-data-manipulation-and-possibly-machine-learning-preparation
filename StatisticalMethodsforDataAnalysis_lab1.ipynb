```python
#i imorted the penguins dataset it contains categoical data
import seaborn as sns
import  matplotlib.pyplot as plt
# first we will imports the dataset
df_penguins = sns.load_dataset('penguins')

print("Successfully loaded the 'penguins' DataFrame!")
```

```python
selected_column = 'species'
```

```python
frequency_distribution = df_penguins[selected_column].value_counts()
print(frequency_distribution)
```

```python
total_values = frequency_distribution.sum()
relative_frequency_distribution = frequency_distribution / total_values
display(relative_frequency_distribution)
```

```python
#Relative frequency distribution
print("Relative Frequency Distribution:")
print(relative_frequency_distribution)
# Draw a bar graph
plt.figure(figsize=(8, 6))
sns.barplot(x=relative_frequency_distribution.index, y=relative_frequency_distribution.values)
plt.title('Relative Frequency Distribution of Penguin Species')
plt.xlabel('Species')
plt.ylabel('Relative Frequency')
```

```python
 # now the Pie chart

import matplotlib.pyplot as plt

plt.figure(figsize=(8, 8))
plt.pie(relative_frequency_distribution, labels=relative_frequency_distribution.index, autopct='%1.1f%%', startangle=140)
plt.title('Relative Frequency Distribution Pie Chart')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
